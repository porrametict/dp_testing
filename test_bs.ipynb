```python
from bigstream import StorageService

ss = StorageService(url="http://datalake.opend.cloud/v1.2",token='')
```

```python
import cProfile, pstats, io
import psutil
from pstats import SortKey
```

```python
pr = cProfile.Profile()
pr.enable()
storages = ss.get_storages()

objs  = storages["ckandata.json.7518f02f-5614-4779-84ab-13f4d687ac11.0f34f996-8c05-4a63-be63-c7d122d48e1c"]

data = objs.read({"from":0,"limit":100000})

s = io.StringIO()
sortby = SortKey.CUMULATIVE
ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
ps.strip_dirs()
ps.print_stats('json')
print(s.getvalue())
data
```
